```python
import pandas as pd 
import seaborn as sns
import matplotlib
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_df.dropna(subset=["RainToday","RainTomorrow"],inplace=True)
```

```python
raw_df.info()
```

```python
import numpy as np
```

```python
numeric_cols = raw_df.select_dtypes(include=np.number).columns.tolist()
numeric_cols
```

```python
categorical_cols = raw_df.select_dtypes(include="object").columns.tolist()
categorical_cols
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "mean")
```

```python
#raw_df[numeric_cols]=imputer.fit_transform(raw_df[numeric_cols])  #returns a numpy array which is plotted to the dataframe
#better way just to make sure we plot to exactly same columns is 
raw_df[numeric_cols] = pd.DataFrame(imputer.fit_transform(raw_df[numeric_cols]),columns=numeric_cols,index=raw_df.index)
```

```python
raw_df[numeric_cols]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
raw_df[numeric_cols] = pd.DataFrame(scaler.fit_transform(raw_df[numeric_cols]),columns=numeric_cols,index=raw_df.index)
raw_df[numeric_cols]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
raw_df[categorical_cols[1:-1]].nunique()  #[1:] i dont want to include date as its too big and want to split data according to date 
```

```python
encoder = OneHotEncoder(sparse_output = False)  #returns a sparse matrix when true
```

```python
encoded_data = encoder.fit_transform(raw_df[categorical_cols[1:-1]])
encoded_data
```

```python
encoded_cols = encoder.get_feature_names_out(categorical_cols[1:-1])
encoded_cols
```

```python
encoded_df = pd.DataFrame(encoded_data,columns=encoded_cols,index=raw_df.index)
encoded_df
```

```python
final_df = pd.concat([raw_df[numeric_cols],encoded_df],axis=1).copy()
final_df   #contains only numeric and encoded cols
```

```python
final_df[["Date"]] = raw_df[["Date"]]  #we are taking date column from raw_df to final_df
final_df[["RainTomorrow"]] = raw_df[["RainTomorrow"]]
final_df  #this is our final data which we will split according to year and further into inputs and targets
```

```python
year = pd.to_datetime(final_df["Date"]).dt.year
year
```

```python
train_df = final_df[year <2015]
val_df = final_df[year == 2015]
test_df = final_df[year > 2015]
```

```python
print("Train data : ",train_df.shape)
print("Val data : ",val_df.shape)
print("Test data :",test_df.shape)
```

```python
X_train = train_df.drop(columns=["RainTomorrow","Date"])
y_train = train_df[["RainTomorrow"]]
X_val = val_df.drop(columns=["RainTomorrow","Date"])
y_val = val_df[["RainTomorrow"]]
X_test = test_df.drop(columns=["RainTomorrow","Date"])
y_test = test_df[["RainTomorrow"]]
```

```python
def playing_with_hyperParameters(**params) :
    print("Parameters passed to DecisionTreeClassifier:", params)
    model = DecisionTreeClassifier(**params,random_state=42)
    model.fit(X_train,y_train)
    print("Actual Tree Depth:", model.tree_.max_depth)
    training_score = model.score(X_train,y_train)
    print("Training Score : ",training_score)
    Validation_score = model.score(X_val,y_val)
    print("Validation Score : ",Validation_score)
```

```python
playing_with_hyperParameters()
```

```python
def max_depth_error(md) :
    model = DecisionTreeClassifier(max_depth = md,random_state=42)
    model.fit(X_train,y_train)
    train_error = 1 - model.score(X_train,y_train)
    val_error = 1- model.score(X_val,y_val)
    return {"max_depth" : md,
            "train_error" : train_error,
            "val_error" : val_error}
```

```python
results =[]
for md in range (1,21) :
    print(results.append(max_depth_error(md)))
```

```python
%%time
df = pd.DataFrame(results)
df
```

```python
plt.plot(df["max_depth"],df["train_error"])
plt.plot(df["max_depth"],df["val_error"])
plt.legend(["Training Error","Validation Error"])
plt.xticks(range(0,21,2))
plt.show()
```

```python
def max_leaf_nodes_error(mn) :
    model = DecisionTreeClassifier(max_leaf_nodes=mn,random_state=42)
    model.fit(X_train,y_train)
    train_error = 1 - model.score(X_train,y_train)
    val_error = 1- model.score(X_val,y_val)
    return {"Leaf_nodes" : mn,
            "train_error" : train_error,
            "val_error" : val_error}
```

```python
result = []
for mn in range (1,10) :
    result.append(max_leaf_nodes_error(2**mn))
```

```python
df = pd.DataFrame(result)
df
```

```python
plt.plot(df["Leaf_nodes"],df["train_error"])
plt.plot(df["Leaf_nodes"],df["val_error"])
plt.xlabel("Lead_nodes")
plt.ylabel("Train/Val Error")
plt.legend(["Training Error","Validation Error"])
plt.show()            #from graph we can say that the value of max_node should lie btw 150 and above i.e 128(7) or 256(8) will do 
```
